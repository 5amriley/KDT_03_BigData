## 1. 대구광역시 전체 및 8개 구, 군별 (중구, 동구, 서구, 남구, 북구, 수성구, 달서구, 달성군) 남녀 비율을 각각의 파이 차트로 구현하세요.
- subplots 를 이용하여 3x3 형태의 총 9개의 subplot을 파이차트로 구현
- gender.csv 파일 사용

```python
import pandas as pd
```

```python
# DataFrame으로 저장
filename = '../../EXAM_BIGDATA/DAY_240126/DATA/gender.csv'
genderDF = pd.read_csv(filename, encoding='utf-8-sig', header=[0])
genderDF
```

```python
daegu_mask = genderDF['행정기관'].str.startswith('대구')
genderDF[daegu_mask]
```

```python
# '대구'에 해당하는 행만 필터링하여 daeguDF에 저장
daeguDF = genderDF[daegu_mask]
daeguDF.iloc[:, [0, 104, 207]]
```

```python
daeguDF2 = daeguDF.iloc[:, [0, 104, 207]].copy()
daeguDF2.info()
```

```python
daeguDF2['남 인구수'] = daeguDF2['남 인구수'].str.replace(',', '')
daeguDF2['여 인구수'] = daeguDF2['여 인구수'].str.replace(',', '')
daeguDF2 = daeguDF2.astype({'남 인구수':'int32', '여 인구수':'int32'})
daeguDF2.info()
```

```python
city_name = []
gender_population = []

for i in range(9):
    city_name.append(daeguDF2.iloc[i, 0].strip())
    gender_population.append([daeguDF2.iloc[i, 1], daeguDF2.iloc[i, 2]])
print(city_name)
print(gender_population)
```

```python
import matplotlib.pyplot as plt
import koreanize_matplotlib
```

```python
colors = ['tomato', 'royalblue']
labels = ['남성', '여성']
```

```python
plt.subplot(2, 1, 1)
plt.pie(gender_population[0], colors=colors, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title(city_name[0])
plt.subplot(2, 1, 2)
plt.pie(gender_population[1], colors=colors, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title(city_name[1])
plt.show()
```

#### plt.subplot() 으로 해결하는 방법

```python
for i in range(9):
    plt.subplot(3, 3, i+1)
    plt.pie(gender_population[i], colors=colors, labels=labels, autopct='%1.1f%%', startangle=90)
    plt.title(city_name[i])
plt.suptitle('대구광역시 구별 남녀 인구 비율')
plt.show()
```

#### plt.subplots() 로 해결하는 방법

```python
# fig => <class 'matplotlib.figure.Figure'>
# ax  => <class 'numpy.ndarray'>
fig, ax = plt.subplots(3, 3)
for i in range(3):
    for j in range(3):
        ax[i, j].pie(gender_population[i*3+j], colors=colors, labels=labels, autopct='%1.1f%%', startangle=90)
        ax[i, j].set_title(city_name[i*3+j])
plt.suptitle('대구광역시 구별 남녀 인구 비율')
plt.show()
```
