## 2. 지하철 각 노선별 퇴근 시간대 최대 하차 인원을 막대그래프로 표시하고, 하차인원 출력  
- 퇴근 시간대 : 19:00~20:59
- 사용 파일 : subwaytime.csv 또는 subway.xls
- 가장 많이 내리는 지하철역 순서로 정렬 후 10개의 지하철역 및 하차 인원 수를 화면에 출력.
- 지하철 호선은 다르지만, 동일한 이름의 역은 모두 합해서 계산함.
- 하차 인원은 1,000 단위로 콤마를 찍어서 구분할 것
- Bar chart의 x축은 지하철 역 이름을 표시하고, y축은 인원수를 표시

```python
import pandas as pd
import numpy as np
```

```python
filename = '../../DATA/subwaytime.csv'
subwayDF = pd.read_csv(filename, encoding='utf-8-sig', header=[0, 1])
subwayDF
```

```python
commute_time_df = subwayDF.iloc[:, [1, 3, 35, 37]]
commute_time_df
```

```python
# commute_time_df column 레이블 재설정
commute_time_df.columns = ['호선명', '지하철역', '19시 중 하차', '20시 중 하차']
commute_time_df
```

```python
# commute_time_df를 바로 바꾸면 Warning 발생하니, copy본을 만들어 수정한다.
# commute_time_df를 정말 바꾸고 싶으면, commute_time_df에 수정이 끝난 copy본을 할당한다.
copyDF = commute_time_df.copy()
copyDF
```

```python
copyDF[('퇴근시간 하차')] = copyDF['19시 중 하차'] + copyDF['20시 중 하차']
copyDF
```

```python
copyDF['지하철역'].nunique()
```

```python
station_list = np.sort(copyDF['지하철역'].unique())
station_list
```

```python
# 각 지하철역별 퇴근시간 하차인원을 저장할 빈 DataFrame 생성
unique_station_df = pd.DataFrame(columns=['지하철역', '퇴근시간 하차'])
unique_station_df.info()
```

```python
# DataFrame에 행 추가할 때 loc[] 이용
idx = 0
for station in station_list:
    station_mask = copyDF['지하철역'] == station
    stationDF = copyDF[station_mask]
    unique_station_df.loc[idx] = [station, stationDF['퇴근시간 하차'].sum()]
    idx += 1
unique_station_df
```

```python
# 퇴근시간 하차인원 기준으로 내림차순으로 정렬
unique_station_df = unique_station_df.sort_values('퇴근시간 하차', ascending=False)
unique_station_df[:5]
```

```python
# 인덱스 초기화
unique_station_df = unique_station_df.reset_index(drop=True)
unique_station_df
```

```python
X = [unique_station_df.loc[i, '지하철역'] for i in range(7)]
Y = [unique_station_df.loc[i, '퇴근시간 하차'] for i in range(7)]
```

```python
unique_station_df.loc[0, '지하철역']
```

```python
for i in range(10):
    print(f"{unique_station_df.loc[i, '지하철역']}: {unique_station_df.loc[i, '퇴근시간 하차']:,}")
```
